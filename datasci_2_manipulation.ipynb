# Data Cleaning with Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('healthcare_data_cleaning.csv')
df
```

```python
data_types= df.dtypes
data_types
```

```python
df ['Payment Due ($)'] = df['Payment Due ($)'].astype(float)
```

## Removing Duplicates

```python
df.drop_duplicates(inplace=True)
df
```

## Removed White Space and Special Characters

```python
import pandas as pd
import re
import numpy as np
```

```python
def clean_value(value):
    cleaned_value = re.sub(r'\s+|[^a-zA-Z0-9]', '', value)
    return cleaned_value

# Apply the clean_value function to all columns
for column in df.columns:
    df[column] = df[column].apply(clean_value)

df
```

## Adding Decimals Back to Payment Due and Presciption Cost Column

```python

import pandas as pd

# Sample DataFrame with cleaned money amounts (without decimals)
data = {'Payment Due ($)': [65.45721578126224, 430.68365678679174, 315.0709305262176, 320.2998987723972, 264.2147372473319]}
df = pd.DataFrame(data)

# Number of decimal places to round
decimal_places = 2

# Function to round numbers to the specified decimal places
def round_to_decimal_places(value, decimal_places):
    return round(value, decimal_places)

# Apply the rounding function to the 'Amount' column
df['Payment Due ($)'] = df['Payment Due ($)'].apply(lambda x: round_to_decimal_places(x, decimal_places))

# Display the corrected DataFrame
df
```

## Handling Missing Values

```python
df = df.replace('missing', 'Unknown')
df
```

## Cleaning Column Names

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
df
```
