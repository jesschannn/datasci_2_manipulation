# Data Cleaning with Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('healthcare_data_cleaning.csv')
df
```

```python
data_types= df.dtypes
data_types
```

## Removing Duplicates

```python
df.drop_duplicates(inplace=True)
df
```

## Removed White Space and Special Characters

```python
import pandas as pd
import re
import numpy as np
```

```python
def clean_value(value):
    cleaned_value = re.sub(r'\s+|[^a-zA-Z0-9]', '', value)
    return cleaned_value

# Apply the clean_value function to all columns
for column in df.columns:
    df[column] = df[column].apply(clean_value)

df
```

## Adding Decimals Back to Payment Due and Presciption Cost Column

```python
import re
import numpy as np
```

```python
df['Payment Due ($)'] = pd.to_numeric(df['Payment Due ($)'], errors='coerce')
df['Prescription Cost ($)'] = pd.to_numeric(df['Prescription Cost ($)'], errors='coerce')
```

```python
df['Patient Age'] = pd.to_numeric(df['Patient Age'], errors='coerce')
df['Average Heart Rate'] = pd.to_numeric(df['Average Heart Rate'], errors='coerce')
df['Average BP'] = pd.to_numeric(df['Average BP'], errors='coerce')
df['Height (in cm)'] = pd.to_numeric(df['Height (in cm)'], errors='coerce')
df['Weight (in kg)'] = pd.to_numeric(df['Weight (in kg)'], errors='coerce')
```

```python
df['Last Visit (days ago)'] = pd.to_numeric(df['Last Visit (days ago)'], errors='coerce')
df['Visit Duration (mins)'] = pd.to_numeric(df['Visit Duration (mins)'], errors='coerce')
df['Number of Tests'] = pd.to_numeric(df['Number of Tests'], errors='coerce')
```

## Handling Missing Values

```python
missing_values = df.isnull().sum()
missing_values
```

```python
df = df.replace('missing', 'Unknown')
df
```

## Cleaning Column Names

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
df
```
