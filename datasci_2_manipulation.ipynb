# Data Cleaning with Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/hantswilliams/HHA_507_2023/main/WK2/data/healthcare_data_cleaning.csv')
df
```

## Removing Duplicates

```python
df.drop_duplicates(inplace=True)
df
```

## Removed White Space and Special Characters

```python
import pandas as pd
import re

def clean_value(value):
    cleaned_value = re.sub(r'\s+|[^a-zA-Z0-9]', '', value)
    return cleaned_value

# Apply the clean_value function to all columns
for column in df.columns:
    df[column] = df[column].apply(clean_value)

df
```

## Handling Missing Values

```python
df = df.replace('missing','Unknown')
df
```

## Cleaning Column Names

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
df
```

## Summary of Changes Made to the Dataset

After loading the dataset into Google Colab, I removed the duplicates first. Removing the duplicates brought the amount of records / rows down from 10,500 to 10,000. Then I removed white space and special characters from the dataset only (excluding the columns). I noticed that there weren't any white spaces in the data under the "City of Residence" column & no more periods and white space in the data under the "Preferred Doctor" column. After that, I identified and handled missing values in the dataset. I noticed that all 20 columns had missing data. I handled the missing data by substituting "missing" with "unknown". Lastly, I cleaned the column names. In the original dataset, there were white spaces in the column names. There are now underscore characters where the white spaces used to be.
